### 6. Wrangle the data in preparation for machine learning

rubric={correctness:20}

Our data currently covers all of NSW, but say that our client wants us to create a machine learning model to predict rainfall over Sydney only. There's a bit of wrangling that needs to be done for that:
1. We need to query our data for only the rows that contain information covering Sydney
2. We need to wrangle our data into a format suitable for training a machine learning model. That will require pivoting, resampling, grouping, etc.

To train an ML algorithm we need it to look like this:

||model-1_rainfall|model-2_rainfall|model-3_rainfall|...|observed_rainfall|
|---|---|---|---|---|---|
|0|0.12|0.43|0.35|...|0.31|
|1|1.22|0.91|1.68|...|1.34|
|2|0.68|0.29|0.41|...|0.57|

6.1) Get the data from s3 (```combined_model_data_parti.parquet``` and ```observed_daily_rainfall_SYD.csv```)

6.2) First query for Sydney data and then drop the lat and lon columns (we don't need them).
```
syd_lat = -33.86
syd_lon = 151.21
```
Expected shape ```(1150049, 2)```.

6.3) Save this processed file to s3 for later use:

  Save as a csv file ```ml_data_SYD.csv``` to ```s3://mds-s3-xxx/output/```
  expected shape ```(46020,26)``` - This includes all the models as columns and also adding additional column ```Observed``` loaded from ```observed_daily_rainfall_SYD.csv``` from s3.

````python
### Do all your coding here
````

````python
## Imports
import re
import os
import glob
import zipfile
import requests
from urllib.request import urlretrieve
import json
import pandas as pd
````

````python
# read the credentials
with open('/srv/data/my_shared_data_folder/aws_credential.json') as f:
    aws_credentials = json.load(f)
````

````python
# read the data for combined parquet file from s3 and filter for Sydney 

df_combined = pd.read_parquet('s3://mds-s3-24/combined_model_data_parti.parquet',
                              filters=[[('lat_min','<=', -33.86),('lat_max', '>=', -33.86),
                                        ('lon_min','<=', 151.21),('lon_max','>=', 151.21)]],
                              storage_options=aws_credentials)

df_combined = df_combined.drop(columns = ['lat_min', 'lat_max', 'lon_min', 'lon_max'])
````

````python
df_combined['time'] = pd.to_datetime(df_combined['time']).dt.date
df_combined = df_combined.set_index('time')
df_combined.head()
````

````python
df_combined.shape
````

````python
# read the data from observed daily for Sydney csv file from s3
# add new column model with value as observed_rainfall


df_observed = pd.read_csv('s3://mds-s3-24/observed_daily_rainfall_SYD.csv',
                              storage_options=aws_credentials)
df_observed["model"] = "observed_rainfall"
df_observed['time'] = pd.to_datetime(df_observed['time']).dt.date
df_observed = df_observed.set_index('time')
df_observed.head()
````

````python
df_observed.shape
````

````python
# Concat two dataframes

df = pd.concat([df_combined, df_observed])
````

````python
df.shape
````

````python
# add models as columns 

df_models = df.reset_index().pivot(index='time', columns='model', values='rain (mm/day)')
````

````python
df_models
````

````python
df_models.shape
````

````python
# Save this processed file to s3 
df_models.to_csv('s3://mds-s3-24/output/ml_data_SYD.csv',
                 storage_options=aws_credentials)
````

````python
# Check the saved file
pd.read_csv('s3://mds-s3-24/output/ml_data_SYD.csv',
                              storage_options=aws_credentials,index_col = 0)
````
